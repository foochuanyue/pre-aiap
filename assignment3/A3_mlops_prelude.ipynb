# 3. Docker Engine

(This section contains basic information on Docker Engine. Additional readings are required to have deeper understanding.)

- “Open-source containerisation tool for building and containerising applications.”
- Can be installed directly on Linux distributions without Docker Desktop.
- [Docker Desktop](https://docs.docker.com/desktop/) is an additional interface for making usage of Docker Engine possible for Windows and macOS machines.
- Docker Desktop is subjected to subscription based on usage.
- For Windows OS, [WSL2](https://docs.microsoft.com/en-us/windows/wsl/install) is needed for usage of GPU within Docker containers.
- [Docker Hub](https://www.docker.com/products/docker-hub/) is a centralised place for storing and sharing Docker images; free users subjected to usage rates and limitations.


# 4.  Miniconda VS Anaconda

- [Anaconda](https://www.anaconda.com/products/distribution) is a distribution platform.
  - Includes a GUI for managing environments and packages.
  - Includes libraries and tools out of the box.
  - Includes the `conda` CLI, a package and virtual environment management tool.

- [Miniconda](https://docs.conda.io/en/latest/miniconda.html) is a minimal installer for `conda`.
  - Recommended for environments where footprint is to be minimised.
  - The go-to installer for containers that requires `conda`.
